```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
avocado_df = pd.read_csv('avocado.csv')
display(avocado_df.head())
```

```python
display(avocado_df.info())
```

```python
avocado_df['Date'] = pd.to_datetime(avocado_df['Date'])
display(avocado_df.info())
```

```python
filtered_df = avocado_df.query("type == 'conventional' and region == 'Albany'")
filtered_df = filtered_df.loc[filtered_df['Date'].dt.year == 2015, ['Date', 'Total Bags', 'Small Bags']]
display(filtered_df.head())
```

```python
melted_df = filtered_df.melt(id_vars=['Date'], value_vars=['Total Bags', 'Small Bags'], var_name='Bags', value_name='Count')
display(melted_df)
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=melted_df, x='Date', y='Count', hue='Bags', ci='sd', estimator='mean')
plt.title('Bags vs. Count Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.show()
```

```python
avocado_df['Sales Volume'] = pd.qcut(avocado_df['Total Volume'], q=4, labels=['Poor', 'Modest', 'Good', 'Excellent'])
display(avocado_df.head())
```

```python
avocado_df['year'] = avocado_df['Date'].dt.year
sns.countplot(data=avocado_df, x='year', hue='Sales Volume')
```

```python
bag_melted_df = avocado_df.melt(id_vars=['Date'], value_vars=['Small Bags', 'Large Bags', 'XLarge Bags'], var_name='Bags', value_name='Count')
display(bag_melted_df)
```

```python
sns.relplot(data=bag_melted_df, x='Date', y='Count', hue='Bags', kind='line', ci='sd')
```
