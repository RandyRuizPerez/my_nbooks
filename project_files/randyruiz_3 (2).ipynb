```python
import pandas as pd
```

## Get the data

```python
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

```python
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head()
```

## Visualize the data

```python
mortality_data.query('AgeGroup == "01-04 Years"').plot.scatter(x='Year', y='DeathRate')
```

```python
mortality_wide.plot()
```

```python
mortality_wide.plot.line(y=['15-19 Years'])
```

```python
mortality_wide.iloc[:,::-1].plot.area()
```

```python
mortality_wide.loc[2000].plot.bar(ylabel='Mortality Rate').set(xlabel=None)
```

```python
mortality_wide.loc[2000].plot.bar(ylabel='Mortality Rate').set(xlabel=None)
```

```python
mortality_data.query(Year =='2000').pivot(index='', columns='AgeGroup', values='DeathRates').plot.bar(title=
'Mortality Rates', ylabel='Mortality Rate', legend=True)
```

```python
mortality_data['DeathRate'].plot.hist(bins=15, title='Frequency of Death Rates')
```

```python
mortality_data['DeathRate'].plot.density(y='1900', grid=True)
```

```python
mortality_data[mortality_data['Year'].isin([1900, 1925, 1950, 1975, 2000])].groupby('Year')['DeathRate'].sum().plot.pie(title='Sum of Death Rates')
```

```python
mortality_wide.plot.line(title=['Child Mortality: 01-04','Child Mortality: 05-09', 'Child Mortality: 10-14','Child Mortality: 15-19'],
                         ylabel='Deaths per 100,000', sharey=True,grid=True, rot=45, xlim=(1900, 2000), legend=False, subplots=True, layout=(2,2), figsize=(10,10))
```
